### 필요 라이브러리 import / 크롬 브라우저 불러오기

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import urllib.request

chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
```

```python
# 구글 이미지 접속

url = 'https://www.mangoplate.com/'
driver.get(url)

# 검색어 입력 및 조회
search_name = '냉면'

# 입력 Box 선택
input_box = driver.find_element(By.CSS_SELECTOR, 'input.HomeSearchInput')
input_box.clear()
input_box.send_keys(search_name)
input_box.send_keys(Keys.ENTER)
time.sleep(3)

#원하는 탭 클릭
driver.find_element_by_xpath('//*[@id="lnb"]/div[1]/div/ul/li[2]/a').click()
```

```python


## 스크롤 다운 
SCROLL_PAUSE_TIME = 1

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            break
    last_height = new_height

driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.HOME) # 스크롤 맨 위로 이동 (몇 초 소요됨)
##

#img_box = driver.find_element(By.CSS_SELECTOR, 'div.islrc')
images = driver.find_elements(By.CSS_SELECTOR, '.rg_i.Q4LuWd')
image_save_folder = 'C:/Users/ADMIN/Documents/수업파일/221214(Semi)/crawling데이터/크롤링된곳/'  # 저장 폴더 지정
count = 1

for image in images:
    try:
        image.click()  # 썸네일 클릭
        time.sleep(3)
        imgUrl = driver.find_element(By.CSS_SELECTOR, 'img.n3VNCb.KAlRDb').get_attribute('src') # 큰이미지 (원본이미지) 주소 가져오기        
        file_name = image_save_folder + search_name + str(count + 1) + ".jpg"
        urllib.request.urlretrieve(imgUrl, file_name)
        print(file_name)
        count += 1
    except:  # 저장이 안되는 데이터는 넘어가기
        pass
print('데이터 수집을 완료하였습니다.')
```

```python
# 구글 이미지 접속

url = 'https://images.google.com/'
driver.get(url)

# 검색어 입력 및 조회

search_name = input('검색하고 싶은 이미지: ')
img_cnt = int(input('원하는 이미지 수집 수: '))

# 입력 Box 선택

input_box = driver.find_element(By.CSS_SELECTOR, 'input.gLFyf')
input_box.clear()
input_box.send_keys(search_name)
input_box.send_keys(Keys.ENTER)
driver.implicitly_wait(10)

## 스크롤 다운 
SCROLL_PAUSE_TIME = 1

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            break
    last_height = new_height

driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.HOME) # 스크롤 맨 위로 이동 (몇 초 소요됨)
##

image_save_folder = 'C:/Users/ADMIN/Documents/수업파일/221214(Semi)/crawling데이터/크롤링된곳/'  # 저장 폴더 지정

for i in range( img_cnt ) :
    try:
        images = driver.find_elements(By.CSS_SELECTOR, '.rg_i.Q4LuWd')[i].click()
        time.sleep(3)
        imgUrl = driver.find_element(By.CSS_SELECTOR, 'img.n3VNCb.KAlRDb').get_attribute('src') # 큰이미지 (원본이미지) 주소 가져오기
        file_name = image_save_folder + search_name + str(i + 1) + ".jpg"
        urllib.request.urlretrieve(imgUrl, file_name)
        print(file_name)
    except: # 저장이 안되는 데이터는 넘어가기
        pass
print('데이터 수집을 완료하였습니다.')
```
